### Input Layer:
The raw input the NN receives. Usually many more neurons in the input layer than the output layer. Each raw input is taken into account and weighted for each neuron of the next hidden layer.
### Hidden Layer:
The layers of neurons inbetween the initial input vector and the final output. The purpose of these layers is to reduce the quantity of neurons into an interpretable output.
### Output Layer:
If a neural network is a function, the output layer is the returned vector. The output layer usually is a smaller and more interpretable set of neurons then the input layer.
### Neuron:
A neuron is a number that is used to inform the dynamics of the network. It is calculated using an activation function that procceses a series of weights and neurons from a previous layer. The neurons of the input layer are the initial inputs, they are not calculated.
### Weight:
Weights in NN are a value that designate the importance of one neuron in the calculation of a neuron in the next layer.
### Activation Function:
Activation functions are used in NNs to standardize neuron values between 0 and 1. Examples include relu and sigmoid.
### Node Map:
Node maps are visualization of the neurons and the connections between them.
### Perceptron:
A perception is the simplest neural network, with a vector input and single output.

### Explain the flow of information through a neural network from inputs to outputs. Be sure to include: inputs, weights, bias, and activation functions. How does it all flow from beginning to end?

Neural networks are a mechanism that reduces complex, high-dimensionality input such as text, photography, audio, or video into a simpler, lower-dimensionality input. For the example of handwritten digits, a NN will assign the brightness of each pixel as a value from 0 to 1, and each pixel's brightness will be a neuron in the input layer. Each neuron in the input layer is multiplied by a weight, summed, applied a bias (a value that adds to or reduces the summed value) and processed through an activation function, which normalizes the next neuron's value to be between 0 and 1. This process is repeated for every neuron outside of the input layer, until the output layer is reached. The output layer will be of a lower dimension and more interpretable than the input layer.

```python
##### Update this Class #####

class Perceptron(object):
    
    def __init__(self, bias = 10, n_iter=1000):
        self.bias = bias
        self.n_iter = n_iter
    
    def __sigmoid(self, x):
        return 1/(1+np.exp(-x))
    
    def __sigmoid_derivative(self, x):
        sx = sigmoid(x)
        return sx(1-sx)

    def fit(self, X, y):
        """Fit training data
        X : Training vectors, X.shape : [#samples, #features]
        y : Target values, y.shape : [#samples]
        """

        # Randomly Initialize Weights
        weights = np.random.rand(X.shape[1], 1)

        for i in range(self.n_iter):
            # Weighted sum of inputs / weights
            weighted_sum = np.dot(X, weights) + self.bias
            # Activate!
            self.activated_output = sigmoid(weighted_sum)
            # Cac error
            error = y - self.activated_output
            # Update the Weights
            adjustments = error*sigmoid_derivative(self.activated_output)
            weights += np.dot(X.T, adjustments)
    def predict(self, X):
        """Return class label after unit step"""
        return self.activated_output
```
